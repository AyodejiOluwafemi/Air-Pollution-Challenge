```python
train = train.loc[:, 'Place_ID X Date':'L3_CH4_solar_zenith_angle']
```

```python
train = train.fillna(train.mean())
```

```python
train.isnull().sum()
```

```python
display(train.head())
```

```python
train.set_index('Place_ID X Date', inplace=True)
```

```python
display(train.head())
```

```python
train['Date'] = pd.to_datetime(train['Date'])
train['weekday'] = train['Date'].dt.weekday.astype(str)
train['year'] = train['Date'].dt.year
train['month'] = train['Date'].dt.month.astype(str)
```

```python
train.drop('Date', inplace=True, axis=1)
```

```python
display(train.head())
```

```python
train['year'] = train['year'].astype(str)
train['month'] = train['month'].astype(str)
train['weekday'] = train['month'].astype(str)
```

```python
train = pd.get_dummies(train, columns=['weekday', 'year', 'month'])
```

```python
display(train.head())
```

```python
train['Place_ID'].value_counts()
```

```python
train.drop('Place_ID', axis=1, inplace=True)
```

```python
display(train.head())
```

```python
# Extract Numpy arrays
features = train.loc[:, 'target_min':'month_4']
```

```python
X = features.values
y = train['target'].values
```

```python
# Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2505)

# scale the data
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaler.fit(X_train, y_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

# import the model
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor(random_state=0, max_depth=1).fit(X_train_scaled, y_train)
```

```python
gbrt.score(X_test_scaled, y_test)
```

```python
y_pred = gbrt.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
import numpy as np
print ("rmse on test is %f " %(np.sqrt(mean_squared_error(y_test,y_pred))))
```
